# EDA 

## Univariant analysis

distributing the data to num and cat

```python
num_set = df.select_dtypes(include=['int64','float64'])
cat_set = df.select_dtypes(include=['object'])
print(f'numerical ; {num_set.columns} \n Categorical : {cat_set.columns}')
```

### A brief overview of the data by cat and num

```python
fig, axes = plt.subplots(7, 2,figsize=(20,25))
axes = axes.flatten()
for i, item in enumerate(cat_set.columns):
    sns.countplot(x = item,palette='tab20',ax = axes[i], data = df)
    axes[i].set_title(f'Count of {item}')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(4, 4,figsize=(20,25))
axes = axes.flatten()
for i, item in enumerate(num_set.columns):
    sns.kdeplot(x = item,palette='tab20',ax = axes[i], data = df)
    axes[i].set_title(f'kde of {item}')
plt.tight_layout()
plt.show()
```

### Temporal Factor Univariant

```python
plt.figure(figsize=(15,6))
sns.countplot(x = 'Year', data = df, order= df['Year'].value_counts().iloc[:5].index , palette= 'tab20' , hue='Month' , hue_order= df['Month'].value_counts().index ).set_title('Most amount of accident in a 5 Year by months')
plt.xticks(rotation = 90)
plt.legend(loc = 'center left' , bbox_to_anchor=(1,0.5))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x = 'Month', data = df, order= df['Month'].value_counts().iloc[:3].index , palette= 'tab20' , hue='Day_of_Week').set_title('Most amount of accident in a 3 Month by Days')
plt.xticks(rotation = 90)
plt.legend(loc = 'center left' , bbox_to_anchor=(1,0.5))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x = 'Day_of_Week', data = df, order= df['Day_of_Week'].value_counts().index , palette= 'tab20' , hue='Time_of_Day').set_title('Most amount of accident in a  Day_of_Week by Time_of_Day')
plt.xticks(rotation = 90)
plt.legend(loc = 'center left' , bbox_to_anchor=(1,0.5))
plt.tight_layout()
plt.show()
```

### Location Characteristics

```python
plt.figure(figsize=(10,6))
sns.barplot(x = df['Country'],y = df['Population_Density'] , palette='viridis' , order=df['Country'].value_counts().index)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = df['Region'],y = df['Population_Density'] , palette='viridis' , order=df['Region'].value_counts().index)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = df['Road_Type'],y = df['Population_Density'],hue =df['Urban/Rural'] , palette='Set2' , showmeans = True )
plt.tight_layout()
plt.show()
```
